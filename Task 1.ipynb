## Question 2

```python
# Create 100000x1 arrays from normal, uniform and exponential distributions.
normal_array = np.random.normal(size=100000)
uniform_array = np.random.uniform(size=100000)
exponential_array = np.random.exponential(size=100000)
```

```python
# Visualize the normal, uniform and exponential distributions.
plt.hist(normal_array, density=True, bins=700, histtype="step", color="blue", label="normal")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="green", label="uniform")
plt.hist(exponential_array, density=True, bins=700, histtype="step", color="red", label="exponential")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## Question 3

```python
x = np.arange(-5, 6, 0.5)
y = np.arange(-5, 6, 0.5)
x, y = np.meshgrid(x, y)
z = x**2 + y**2
```

```python
figure = plt.figure()
ax = figure.add_subplot(projection='3d')
ax.plot_surface(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Read csv file and prepare data with selected features.
df = pd.read_csv("Pokemon.csv", index_col=0, encoding="latin")
selected_features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
selected_data = df[selected_features]
```

```python
# Calculate pearson standard coefficient and spearman rank correlation.
pearson_coefficient = selected_data.corr(method="pearson")
spearman_correlation = selected_data.corr(method="spearman")
```

```python
# Visualize the results with heatmaps.
sns.heatmap(pearson_coefficient, annot=True)
plt.title("Pearson Standard Coefficient")
plt.figure()
plt.show()

sns.heatmap(spearman_correlation, annot=True)
plt.title("Spearman Rank Correlation")
plt.figure()
plt.show()
```
